```python
# Rewritten script respecting PEP8 conventions
from datetime import datetime


class Patient:
    
    def __init__(self, first_name:str, last_name:str, date_of_birth:datetime):
        self.first_name = first_name
        self.last_name = last_name
        self.date_of_birth = date_of_birth

        
class PatientSample:   
    
    def __init__(self, sample_id:str, patient:Patient, source_tissue:str, 
            extract_method:str, date_of_extraction:datetime, storage_temperature:float, 
            is_processed:bool, has_to_be_returned:bool):

        self.sample_id = sample_id
        self.patient = patient
        self.source_tissue = source_tissue
        self.extract_method = extract_method
        self.date_of_extraction = date_of_extraction
        self.storage_temperature = storage_temperature
        self.is_processed = is_processed
        self.has_to_be_returned = has_to_be_returned
        
    def can_be_returned(self):
        
        """
        Method that check if a sample can be returned 
        to provider or not.
        :return: bool
        """
        
        if (self.has_to_be_returned
            and self.is_processed):
            return True
        return False
      
        
class Fridge:
    
    def __init__(self, min_temperature, max_nb_samples, samples_stored=[]):
        self.min_temperature = min_temperature
        self.max_nb_samples = max_nb_samples
        self.samples_stored = samples_stored       
    
    def add_sample(self, sample):
        
        """
        Add sample to the frige
        """
        
        self.samples_stored.append(sample)

    def is_storage_full(self):
        
        """
        Method checking if the fridge is full
        and returns a boolean.
        :return: bool
        """
        
        return len(self.samples_stored) >= self.max_nb_samples

    def get_samples_that_can_be_returned(self):
        
        """
        Method returning the samples that can
        be returned to the provider
        :return: list[PatientSample]
        """
        
        samples_to_return = []
        for sample in self.samples_stored:
            if sample.can_be_returned():
                samples_to_return.append(sample.sample_id)
                
        return samples_to_return
            
    
```
