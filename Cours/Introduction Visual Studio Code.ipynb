# VSCode

## Qu'est-ce que c'est?

Visual Studio Code (VSCode) est un environnement de développement intégré (EDI) gratuit, open source et multiplateforme développé par Microsoft. 

Un environnement de développement c'est un ensemble d'outils qui aide le développeur à produire du code. Il va contenir un éditeur de texte destiné à la pogrammation, d'un interpréteur ou compilateur pour l'exécution du programme et d'un débogueur pour exécuter le code ligne par ligne

## Pourquoi utiliser

### Création d'applications structurées facilitées
VSCode vous facilitera la création et la hiérarchisation de votre code en différents dossiers et fichiers car vous n'aurez plus besoin de naviguez dans les dossiers de votre code, vous pourrez simplement ouvrir un explorateur qui sera à côté de votre éditeur de texte. 

### Exécution et débogage efficaces
VSCode permet d'exécuter votre code avec l'interpréteur de votre choix et offre des outils de débogage avancés. Le débogage va vous permettre d'examiner et tester le code ligne par ligne, au lieu de parsemer votre code de print.

### Prise en charge de multiples langages
Vous pouvez coder en Python, HTML, CSS, JavaScript, SQL, Java, et bien plus encore avec VSCode. Il suffit de configurer l'interpréteur ou le compilateur approprié.

### Extensions pour plus de fonctionnalités
Augmentez les capacités de VSCode en ajoutant des extensions. Par exemple, utilisez des auto-formatteurs pour respecter les conventions d'écriture ou ajoutez des outils comme GitHub Copilot pour obtenir des suggestions de code basées sur votre code ou vos commentaires.

### Travail collaboratif
L'intégration de Git à VSCode, qui est un logiciel de gestion des gestions de votre code, facilite le suivi des modifications, la collaboration avec d'autres développeurs et la gestion de différentes versions de votre code.

### Conclusion
VSCode est un EDI largement adopté en milieu professionnel. Il offre des outils essentiels pour coder selon les standards industriels.

## Configuration de VSCode pour Python

### Configuration de settings.json (Utilisateur):

On va configurer le fichier settings.json utilisateur qui va permettre de définir des préférence pour la globalité de l'éditeur VSCode. Il est aussi possible de créer des préférences pour un projet spécifiquement mais nous n'allons pas le voir.

- Lancez Visual Studio Code
- Appuyez sur Ctrl+Shift+P pour ouvrir la palette de commande
- Tapez "settings.json"
- Sélectionnez "Preferences: Open user settings (JSON)

Dans ce fichier json on va pouvoir ajouter des lignes afin de personaliser notre éditeur VSCode. Ce que l'on va faire c'est lui donner un interpréteur Python à utiliser par défaut et aussi lui préciser quel terminal utiliser.

#### Spécifier l'interpréteur Python:
- Ouvrez un terminal
- Ouvrez le mode interactif de Python (tapez "python")
- Tapez les lignes suivantes:
    - import sys
    - print(sys.executable)
- Vous devriez obtenir le path de votre interpréteur Python
- Retour à VSCode dans le fichier settings.json, ajoutez la ligne suivante entre les accolades, , en modifiant <vote_path> par le path obtenu antérieurement:
    - "python.defaultInterpreterPath": "votre_path",

#### Spécifier le terminal par défaut (Windows uniquement)
- Dans le fichier settings.json, ajoutez la ligne suivante entre les accolades:
    - "terminal.integrated.defaultProfile.windows": "Command Prompt",

# Exploration du débogage

## Qu'est-ce que c'est le débogage?
Le débogage est le processus par lequel les développeurs identifient, isolent et corrigent les erreurs ou les bugs dans un programme. 
Il permet aussi de comprendre le comportement d'un programme et de le contrôler pendant son exécution (par exemple en modifiant les valeurs des variables).
Durant ce processus, on utilise un outil appelé débogueur qui va éxecuter le programme pas à pas.

## Comment est-ce qu'on débogue?
<b>Points d'arrêts</b> <br/>
Les points d'arrêt sont des marqueurs que vous placez dans votre code source pour indiquer où l'exécution du programme doit être interrompue. 
Lorsque le point d'arrêt est atteint pendant le débogage, le programme s'arrête, et vous pouvez inspecter les valeurs des variables, la pile des appels, et plus encore.



Image Alt Text

On clique sur la marge gauche de l'éditeur pour placer un point d'arrêt sur une ligne spécifique.

<b>Exécution en mode debug</b> <br/>
Pour commencer le débogage, vous devez d'abord lancer votre programme en mode debug. Sur VSCode on peut le faire en cliquant sur le bouton "Run and Debug" dans la barre d'outils ou en appuyant sur F5.

Image Alt Text

<b>Contrôles de débogage</b> <br/>

Image Alt Text

Une fois que le programme est lancé en mode debug, vous pouvez utiliser les contrôles de débogage pour exécuter le programme pas à pas.
* F5 : Continuer l'exécution du programme jusqu'au prochain point d'arrêt.
* F10 : Exécuter le code ligne par ligne.
* F11 : Exécuter le code ligne par ligne, en entrant dans les fonctions.
* Shift + F11 : Exécuter le code ligne par ligne, en sortant des fonctions.
* Ctrl + Shift + F5 : Redémarrer le débogage.
* Shift + F5 : Arrêter le débogage.

## Fenêtre de débogage

Image Alt Text

<b>Variables</b> <br/>
Cette fenêtre affiche l'ensemble des variables locales et globales du programme. Vous pouvez cliquer sur les variables pour les inspecter.
Elle va contenir les variables locales et globales du programme que ce soit des objets, des fonctions, des classes etc...  

Note: Les variables globales sont accessibles dans toutes les fonctions/méthodes du programme, tandis que les variables locales sont accessibles uniquement dans la nction/méthode où elles sont déclarées.

<b>Watch</b> <br/>
Cette fenêtre affiche les variables que vous avez ajoutées à la liste de surveillance. Vous pouvez cliquer sur les variables pour les inspecter.
C'est très utile lorsqu'il y a énormément de variables dans le programme et que vous voulez vous concentrer sur quelques-unes en particulier.

<b>Callstack</b> <br/>
La "Callstack" ou pile d'appels vous montre la pile d'exécution actuelle de votre programme. 
Cela comprend la fonction actuellement en cours d'exécution, la fonction qui l'a appelée, et ainsi de suite, jusqu'au début du thread d'exécution. 
Cela vous aide à comprendre le flux d'exécution de votre programme et à identifier comment vous êtes arrivé au point d'exécution actuel.
