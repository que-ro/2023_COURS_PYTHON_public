# Interpréteur Python

## Qu'est ce qu'un interpréteur Python ?
De manière très simplifiée, un interpréteur Python est un programme qui lit et exécute du code Python.
Il va permettre la traduction du code Python, compréhensible par nous les développeurs, en langage machine, compréhensible par l'ordinateur.

## Pourquoi Python a-t-il besoin d'un interpréteur ?
Les ordinateurs ne comprennent qu'un langage très basique, connu sous le nom de "langage machine" qui est très éloigné du langage que nous parlons et comprenons.
Le langage machine est une suite d'instructions binaires, c'est à dire des 0 et des 1. C'est le seul langage que l'ordinateur peut comprendre.
Le code Python est écrit dans un langage de haut niveau, c'est à dire un langage proche de notre langage naturel. Cependant, pour qu'un ordinateur exécute nos instructions écrites en Python, il faut que celles-ci soient traduites en langage machine. C'est le rôle de l'interpréteur Python.

## Qu'est-ce qui se passe lorsque l'on exécute un script Python ?
### 1. Analyse et compilation du code Python
L'interpréteur va d'abord analyser et compiler le code Python. Il va lire notre script ligne par ligne et vérifier que le code est valide. Si le code est valide, l'interpréteur va alors traduire le code Python en bytecode, un langage intermédiaire entre le code Python et le langage machine.
### 2. Interprétation du bytecode
L'interpréteur va ensuite traduire le bytecode en langage machine. Maintenant c'est un code compréhensible par l'ordinateur.
### 3. Exécution du code machine
Enfin, l'ordinateur va exécuter le code machine traduit.

# Utilisation de l'interpréteur Python
On va maintenant voir comment utiliser l'interpréteur Python. Pour cela, il existe deux modes d'utilisation: le mode exécution de script et le mode interactif.

## Mode exécution de script
Dans un terminal, on pourra exécuter notre script Python en tapant la commande suivante: "python script.py". Cela va exécuter le script Python contenu dans le fichier script.py.

### A vous d'essayer:
1. Dans le dossier de votre choix, créez un fichier script.py et copiez le code suivant: print("Je suis un script Python !")
2. Sauvegardez le fichier. 
3. Ouvrez un terminal et naviguez vers le dossier contenant le fichier script.py.

Pour naviguer dans un terminal, on utilise les commandes suivantes pour Windows:
- cd: permet de changer de dossier
- dir: permet de lister les fichiers et dossiers dans le dossier courant

et les commandes suivantes pour Mac/Linux:
- cd: permet de changer de dossier
- ls: permet de lister les fichiers et dossiers dans le dossier courant

4. Ensuite, tapez la commande suivante: "python script.py". Vous devriez voir apparaître le message "Je suis un script Python !".

aux développeurs d'accéder à des variables et des fonctions liées à l'interpréteur.
